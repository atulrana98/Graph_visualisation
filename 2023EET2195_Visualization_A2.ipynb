```python
G = ig.Graph.Read("Graph_visualisation/email_univ.txt","edges",directed=True)
```

```python
# Convert igraph graph to NetworkX graph
nx_graph = nx.Graph(G.get_edgelist())

# Create the spring layout using NetworkX
layout1 = nx.spring_layout(nx_graph)

# Plot the graph with the spring layout using NetworkX
plt.figure(figsize=(10, 8))
nx.draw(nx_graph, layout1, with_labels=False, node_size=10, node_color='red', font_size=10, font_color='black', font_weight='bold', edge_color='gray')
plt.title('Spring Layout (NetworkX)')
plt.show()
```

```python
# Create the spiral layout using NetworkX
layout2 = nx.spiral_layout(nx_graph)

# Plot the graph with the spiral layout using NetworkX
plt.figure(figsize=(10, 8))
nx.draw(nx_graph, layout2, with_labels=False, node_size=10, node_color='green', font_size=10, font_color='black', font_weight='bold', edge_color='black')
plt.title('Spiral Layout (NetworkX)')
plt.show()
```
